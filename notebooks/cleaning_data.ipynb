```python
def get_patients(path):
    patients = []

    with open(path) as f:
        lines = [line.strip() for line in f]

    current_values = []

    for line in lines:
        parts = line.split()

        if not parts:
            continue

        if "name" in parts:

            current_values.extend(parts)

            patients.append(current_values)

            current_values = []

        else:
            current_values.extend(parts)
    return patients
```

## 2/ Standardize Missing Values

```python
import numpy as np

df.replace("?", np.nan, inplace=True)
df.replace("NA", np.nan, inplace=True)
df.replace("-9", np.nan, inplace=True)

df.isna().sum()
```

## 3/ Convert Data Types

```python
df = df.apply(pd.to_numeric, errors="ignore")
df.info()
```

## 4/ Remove Irrelevant Columns

### Each feature was assessed for relevance, and variables that did not provide meaningful information for identifying heart disease were discarded.

### These features were not considered, as they do not provide meaningful information in the context of the original dataset.

#### restckm, exerckm, thalsev, thalpul, earlobe, lvx1, lvx2, lvx3, lvx4, lvf, cathef, junk, dummy, id, ccf, name

### These variables were excluded from the analysis. Although they may carry some significance, they should not be used as primary determinants since additional factors such as genetics and other underlying conditions must also be considered.

#### smoke, cigs, years, dm, famhist

### Certain clinically relevant features were considered but ultimately omitted from the model to avoid multicollinearity, overfitting, and other issues associated with high-dimensional datasets.

#### pncaden, ekgmo, ekgday, ekgyr, dig, prop, nitr, pro, diuretic, proto, thaltime, trestbpd, rldv5, rldv5e, cmo, cday, cyr, painloc, painexer, relrest, trestbps, htn, thaldur, met, thalach, tpeakbps, tpeakbpd, xhypo, restef, restwm, exeref, exerwm, lmt, ladprox, laddist, diag, cxmain, ramus, om1, om2, rcaprox, and rcadist.

#### The selected 13 features: age, sex, cp (chest pain type), trestbps (resting blood pressure), chol (serum cholesterol), fbs (fasting blood sugar), restecg (resting ECG results), thalach (maximum heart rate achieved), exang (exercise-induced angina), oldpeak (ST depression), slope (ST segment slope), ca (number of major vessels colored by fluoroscopy), and thal (thalassemia status), were retained due to their well-established clinical relevance in predicting coronary artery disease. These features capture demographic information, cardiovascular risk factors, symptom presentation, exercise response, and imaging-based cardiac assessments, and have been validated across multiple studies as significant predictors of heart disease. Their inclusion aligns with standard practice in cardiovascular research and ensures that the model is both interpretable and grounded in clinical evidence. Of course without forgetting the predicted attribute.



```python
relevant_columns = [
    "age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach",
    "exang", "oldpeak", "slope", "ca", "thal", "num"
]

df= df[relevant_columns]
df
```

## 5/Handle Missing Data

```python
missing_percent = df.isna().mean() * 100

print(missing_percent.sort_values(ascending=False))

cols_to_drop = missing_percent[missing_percent > 30].index
df = df.drop(columns=cols_to_drop)

numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

categorical_cols = df.select_dtypes(include=["object", "category"]).columns
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

## 6/Remove Duplicate Rows

```python
df = df.drop_duplicates()
```
