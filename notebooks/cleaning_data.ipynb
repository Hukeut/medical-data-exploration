```python
def get_patients(path):
    patients = []

    with open(path) as f:
        lines = [line.strip() for line in f]

    current_values = []

    for line in lines:
        parts = line.split()

        if not parts:
            continue

        if "name" in parts:

            current_values.extend(parts)

            patients.append(current_values)

            current_values = []

        else:
            current_values.extend(parts)
    return patients
```

## 2/ Standardize Missing Values

```python
import numpy as np

df.replace("?", np.nan, inplace=True)
df.replace("NA", np.nan, inplace=True)
df.replace("-9", np.nan, inplace=True)

df.isna().sum()
```

## 3/ Convert Data Types

```python
df = df.apply(pd.to_numeric, errors="ignore")
df.info()
```

## 4/ Remove Irrelevant Columns

```python
irrelevant_cols = ["id", "ccf", "dummy", "lvx1", "lvx2", "lvx3", "lvx4", "lvf", "cathef", "junk", "name", "thalsev", "thalpul", "earlobe", "restckm", "exerckm"]
df.drop(columns=irrelevant_cols, inplace=True)
df
```

## 5/Handle Missing Data

```python
missing_percentage = (df.isna().sum() / len(df)) * 100
missing_percentage.sort_values(ascending=False)
```
