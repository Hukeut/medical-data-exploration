```python
with open(hungarian_path) as f:
    lines = [line.strip() for line in f]

patients = []
current_values = []

for line in lines:
    parts = line.split()

    # skip empty lines
    if not parts:
        continue

    # check if this line signals end of a patient
    if "name" in parts:

        # add remaining numbers to current patient
        current_values.extend(parts)

        # save patient
        patients.append(current_values)

        # reset for next patient
        current_values = []

    else:
        # just add numbers
        current_values.extend(parts)
```
